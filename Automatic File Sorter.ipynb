```python
import os, shutil
```

```python
path = r"C:/Users/excad/OneDrive/Documents/New folder/"
```

```python
file_name = os.listdir(path)
```

```python

folder_names = ['documents', 'pictures', 'videos']

for loop in range(0,3):
    if not os.path.exists(path + folder_names[loop]):
        os.makedirs(path + folder_names[loop])


```

```python
for file in file_name:
    if ".pdf" in file and not os.path.exists(path + "documents/" + file):
        shutil.move(path + file, path + "documents/" + file)

    elif ".png" in file and not os.path.exists(path + "pictures/" + file):
            shutil.move(path + file, path + "pictures/" + file)

    elif ".mp4" in file and not os.path.exists(path + "videos/" + file):
            shutil.move(path + file, path + "videos/" + file)

    else:
        print("It is not included")
        
```
