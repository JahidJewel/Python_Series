```python
name = input("Write your name: ")
weight = int(input("Enter your weight in pounds: "))
height = int(input("Enter your height in inches: "))
BMI = (weight * 703) / (height * height)
print(BMI)


if BMI>0:
    if(BMI<18.5):
        print(name + ", You are Underweight")
    elif (BMI<24.9):
        print(name + ", You are Normal")
    elif (BMI<29.9):
        print(name + ", You are Overweight. You need to excercise at least 30 mins a day")
    elif (BMI<34.9):
        print(name + ", You are Obese. Spend 1 hour at gym everyday. No excuse")
    elif (BMI<39.9):
        print(name + ", You are Severely Obese. You need a persoanl trainer and a lot of dedication.")
    elif (BMI>40):
        print(name + ", You are Morbidly Obese. Consult with your doctor first, make a plan and keep working on it to stay alive.")
    else:
        print("Enter Valid input: ")
        
```
